```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

```python
plt.figure(figsize=(16, 6))
sns.countplot(x='state', data=df, )
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x='job_openings', y='Industry', data=df, estimator=sum, color='#002E5D' )
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.lineplot(x='founded', y='growth_percentage', data=df,linewidth=4, color='black')
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(x='employees', y='estimated_revenues', data=df,hue='employees' )
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(x='founded', data=df,color='#002E5D')
plt.show()
```
